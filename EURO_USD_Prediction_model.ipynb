```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn import svm
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
import seaborn as sns
from sqlalchemy import create_engine
```

```python
# Define the date range
start_date = '2010-01-01'
end_date = '2024-05-13'
```

```python
# Define the ticker
eur_usd_ticker = 'EURUSD=X'  # Yahoo Finance ticker for EUR/USD exchange rate
```

```python
#Fetch historical data for EUR/USD exchange rate
try:    
    eur_usd_data = yf.download(eur_usd_ticker, start=start_date, end=end_date)
except Exception as e:
    print("Error fetching data:", e)
    exit()
```

```python
# Drop any rows with missing data in the interest rate data
eur_usd_data.dropna(inplace=True)
```

```python
# Display the data
print("EUR/USD Exchange Rate Data:")
eur_usd_data
```

```python
# Create a SQLAlchemy engine
engine = create_engine('sqlite:///financial_data.db')
```

```python
# Store data in SQL database
eur_usd_data.to_sql('eur_usd_data', engine, if_exists='replace', index=True)
```

```python
#Verify data stored in SQL database
with engine.connect() as conn:
    result = conn.execute("SELECT * FROM eur_usd_data LIMIT 5").fetchall()
    print("EUR/USD Data from SQL:")
    for row in result:
        print(row)
```

Data Preparation and Cleaning

```python
# Selecting features and target variable
eur_usd_data['Return'] = eur_usd_data['Close'].pct_change()
eur_usd_data.dropna(inplace=True)
```

```python
eur_usd_data.head()
```

```python
# Features
X = eur_usd_data[['Open', 'High', 'Low', 'Volume', 'Return']]
```

```python
# Target variable
y = eur_usd_data['Close']
```

```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# Normalize and standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Model Initialization and Training
svr = SVR(kernel='rbf')
svr.fit(X_train_scaled, y_train)
```

```python
# Model Evaluation
y_pred = svr.predict(X_test_scaled)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print(f"Initial SVR Model R-squared: {r2}")
print(f"Mean Squared Error: {mse}")
```

```python
from sklearn.inspection import permutation_importance

result = permutation_importance(svr, X_test_scaled, y_test, n_repeats=10, random_state=1) 
feature_importance = result.importances_mean
print(feature_importance)
```

```python
# Data Visualization
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel('Actual EURO/USD Prices')
plt.ylabel('Predicted EURO/USD Prices')
plt.title('Actual vs Predicted EURO/USD Prices')
plt.grid(True)
plt.show()
```

MODEL OPTIMAZATION

Model Initialization, Training, and Evaluation

```python
# Initialize the SVM model
svr = SVR(kernel='rbf')
```

```python
# Define the parameter grid for hyperparameter tuning
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [0.001, 0.01, 0.1, 1],
    'epsilon': [0.1, 0.2, 0.5, 0.3]
}
```

```python
# Initialize GridSearchCV for optimization
grid_search = GridSearchCV(estimator=svr, param_grid=param_grid, cv=5, scoring='r2')
grid_search.fit(X_train_scaled, y_train)
```

```python
# Best model after optimization
best_svr_model = grid_search.best_estimator_
print(f"Best parameters: {grid_search.best_params_}")
```

```python
# Predict and evaluate the optimized model
y_pred = best_svr_model.predict(X_test_scaled)
r2 = r2_score(y_test, y_pred)
print(f"Optimized SVR Model R-squared: {r2}")
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

Data Visualization

```python
# Plot actual vs predicted prices for the optimized model
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel('Actual EURO/USD Prices')
plt.ylabel('Predicted EURO/USD Prices')
plt.title('Actual vs Predicted EURO/USD Prices')
plt.grid(True)
plt.show()
```

```python
# Documenting the optimization process
results = pd.DataFrame(grid_search.cv_results_)
results.to_csv('svm_model_optimization_results.csv', index=False)
```

```python
# Define the R-squared values for each model
models = ['Initial SVR Model', 'Optimized SVR Model']
r_squared_values = [0.61, 0.94]  # Example R-squared values for each model

# Plot bar chart for model comparison
plt.figure(figsize=(10, 6))
plt.bar(models, r_squared_values, color=['blue', 'orange'])
plt.xlabel('Models')
plt.ylabel('R-squared')
plt.title('Model Comparison based on R-squared')
plt.ylim(0, 1)  # Set y-axis limits to ensure consistency
plt.grid(axis='y')
plt.show()
```
